Task is to predict the per capital income in 2020.

(Total income of a country÷Total population of a country) x 100%

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
# df.head(3)
df
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capital income($US)')
plt.scatter(df['year'], df['per capita income (US$)'], color='green', marker='+')
```

```python
df['per capita income (US$)'] = pd.to_numeric(df['per capita income (US$)'])

x = df['year'].values.reshape(-1,1) # dependent variable as x
y = df['per capita income (US$)'].values # independent variable as y

reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```
